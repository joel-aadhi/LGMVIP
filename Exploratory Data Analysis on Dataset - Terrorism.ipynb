```python
data.rename(columns={'iyear':'Year','imonth':'Month','iday':"day",'gname':'Group','country_txt':'Country','region_txt':'Region','provstate':'State','city':'City','latitude':'latitude',
    'longitude':'longitude','summary':'summary','attacktype1_txt':'Attacktype','targtype1_txt':'Targettype','weaptype1_txt':'Weapon','nkill':'kill',
     'nwound':'Wound'},inplace=True)
```

```python
data = data[['Year','Month','day','Country','State','Region','City','latitude','longitude',"Attacktype",'kill',
               'Wound','target1','summary','Group','Targettype','Weapon','motive']]
```

```python
data.shape
```

## 4. EXPLORATORY DATA ANALYSIS

```python
data.isnull().sum()
```

```python
data['Wound'] = data['Wound'].fillna(0)
data['kill'] = data['kill'].fillna(0)
```

```python
data['Casualities'] = data['kill'] + data['Wound']
```

```python
data.info()
```

```python
year = data['Year'].unique()
years_count = data['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (18,10))
sns.barplot(x = year,
           y = years_count)
plt.xticks(rotation = 50)
plt.xlabel('Attacking Year',fontsize=20)
plt.ylabel('Number of Attacks Each Year',fontsize=20)
plt.title('Attacks In Years',fontsize=25)
plt.show()
```

---

```python
pd.crosstab(data.Year, data.Region).plot(kind='area',stacked=False,figsize=(20,10))
plt.title('Terrorist Activities By Region In Each Year',fontsize=25)
plt.ylabel('Number of Attacks',fontsize=20)
plt.xlabel("Year",fontsize=20)
plt.show()
```

---

```python
attack = data.Country.value_counts()[:10]
attack
```

```python
plt.subplots(figsize=(20,10))
sns.barplot(data['Country'].value_counts()[:10].index,data['Country'].value_counts()[:10].values)
plt.title('Top Countries Affected',fontsize=25)
plt.xlabel('Countries',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.xticks(rotation = 50)
plt.show()
```

---

```python
df = data[['Year','kill']].groupby(['Year']).sum()
fig, ax4 = plt.subplots(figsize=(20,10))
df.plot(kind='bar',alpha=0.8,ax=ax4)
plt.xticks(rotation = 50)
plt.title("People Died By Attack",fontsize=25)
plt.ylabel("Number of people killed",fontsize=20)
plt.xlabel('Year',fontsize=20)
top_side = ax4.spines["top"]
top_side.set_visible(False)
right_side = ax4.spines["right"]
right_side.set_visible(False)
```

---

```python
data[['Attacktype','kill']].groupby(["Attacktype"],axis=0).sum().plot(kind='bar',figsize=(20,10), color='red')
plt.xticks(rotation=50)
plt.title("BAR GRAPH SHOWING THE NUMBER OF PEOPLE KILLED ",fontsize=20)
plt.ylabel('Number of people',fontsize=15)
plt.xlabel('Attack type',fontsize=15)
plt.show()
```

---

```python
data['Attacktype'].value_counts().plot(kind='bar',figsize=(10,10),color='BLACK')
plt.xticks(rotation = 50)
plt.xlabel("Type of Attack",fontsize=15)
plt.ylabel("Number of Attacks",fontsize=15)
plt.title("NAMES OF TYPE OF ATTACK",fontsize=20)
plt.show()
```

---

```python
data[['Group','kill']].groupby(['Group'],axis=0).sum().drop('Unknown').sort_values('kill',ascending=False).head(10).plot(kind='bar',figsize=(10,10))
plt.title("Top 10 Terrorist Group Attacks",fontsize=20)
plt.xlabel("Terrorist Group Name",fontsize=15)
plt.ylabel("Number of People Killed",fontsize=15)
plt.show()
```

---

```python
data['City'].value_counts().to_frame().sort_values('City',axis=0,ascending=False).head(10).plot(kind='bar',figsize=(10,10))
plt.xticks(rotation = 50)
plt.xlabel("City",fontsize=10)
plt.ylabel("Number of attack",fontsize=10)
plt.title("Top 10 most effected city",fontsize=10)
plt.show()
```
